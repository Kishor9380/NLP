##**TASK-1**

Task 1: Conceptual Understanding
1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" can be treated as two different words because of the difference in capitalization. For example, "Love" might appear at the beginning of a sentence or be used as a name, while "love" is usually used as a regular word.

If we don’t standardize the text (like converting everything to lowercase), the model may think they are unrelated, which can create unnecessary confusion. That’s why most NLP tasks include converting text into lowercase so that words like these are treated the same.

2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain a lot of common words like is, the, and, in, which usually don’t add much meaning. This can make the data noisy and increase the size of the dataset.

As a result, the model may take longer to process the data and might focus on less important words instead of meaningful ones. However, in some cases, keeping stopwords is useful because they help maintain the context of a sentence.

3. Provide two real-world scenarios where removing stopwords can be harmful.

a. Sentiment Analysis:
In a sentence like "I do not like this movie", the word "not" is very important. If we remove it, the sentence becomes "I like this movie", which completely changes the meaning.

b. Chatbots or Question Systems:
When someone asks a question like "What is the best way to learn coding?", removing stopwords can make the sentence unclear. The structure of the question gets lost, which can confuse the system and lead to incorrect answers.

4. Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both used to reduce words to their base form, but they work differently.

Stemming is a simple method where parts of words are chopped off. It doesn’t always give a real word. For example, "studies" might become "studi".

Lemmatization, on the other hand, is more advanced. It considers the meaning and context of the word and converts it into a proper base form. For example, "better" becomes "good".

In short, stemming is faster but less accurate, while lemmatization is slower but gives better and more meaningful results.

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove non-alphabetic characters (keep emojis removed)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short tokens (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
# def preprocess_text(text):
#   text=text.lower()

#   #Remove punctuations
#   text=text.translate(str.maketrans('','',string.punctuation))

#   #Remove special Characters/Emojs
#   text=re.sub(r'[^\w\s]','',text)

#   #Tokenization
#   tokens=text.split()

#   #Stopwords removal
#   tokens=[word for word in tokens if word not in stop_words]

#   #Lemmatization
#   tokens=[lemmatizer.lemmatize(word) for word in tokens]

#   return tokens
```

**TASK-4**

```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0.0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_token_length, 2)
results = []

for i in text:
    tokens, cleaned = preprocess_text(i)
    total, unique, avg_len = token_analytics(tokens)

    results.append({
        "original": i,
        "cleaned": cleaned,
        "tokens": tokens,
        "total_tokens": total,
        "unique_tokens": unique,
        "avg_token_length": avg_len
    })
for r in results:
    print("Original:", r["original"])
    print("Cleaned:", r["cleaned"])
    print("Tokens:", r["tokens"])
    print("Total Tokens:", r["total_tokens"])
    print("Unique Tokens:", r["unique_tokens"])
    print("Avg Token Length:", r["avg_token_length"])
    print("-" * 50)
```

```python
def tokens_to_sentence(tokens):
  return ' '.join(tokens)
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
